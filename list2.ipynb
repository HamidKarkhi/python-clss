# WAP to input 5 number and store them in a list if the number is even

```python
even_numbers = []

for i in range(5):

    num = int(input(f"Enter number {i+1}: "))

    if num % 2 == 0:
        even_numbers.append(num)

print("even number:", even_numbers)
```
