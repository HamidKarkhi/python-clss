```python
squre = lambda x:x**2
print(squre(7))
```

```python
even_odd = lambda n : "Even" if n%2==0 else "odd"
print(even_odd(7))
```

```python
lst = [1,2,3,4,5,6,7] 
new_lst = []
for i in list:
    new_lst.append(i**2)
print(new_lst)
```

```python
squares = list(map(lambda x:x**2,lst))
print(squares)
```
