```python
text = "madam"

reverse = ""

for ch in text:
    reverse = ch + reverse

print(reverse)

if text == reverse:
    print("Palindrome")
```

```python
num = 153

temp = num
total = 0

while temp > 0:
    digit = temp % 10
    total += digit ** 3
    temp //= 10

if total == num:
    print("Armstrong")
else:
    print("Not Armstrong")
```
