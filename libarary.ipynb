```python
import os
```

```python
print("Get current working diresctory", os.getcwd())
```

```python
print("List all the directories and files in current working directory", os.listdir())
```

```python
total_folders = 0
total_files = 0
for path, folders, files in os.walk(os.getcwd()):
    print("Currents Path:", path)
    print("Folders:", folders)
    print("Files:", files)
    total_folders += len(folders) 
    total_files += len(files)
print("Total Folders:", total_folders)
print("Total_files:", total_files)
```
