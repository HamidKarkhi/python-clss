```python
file = open("data.txt", "r")
print(file.name) #name o the file
print(file.mode) #mode in which the file is opened
print(file.closed) # check if the file is closed or not
file.read(5) 
print(file.tell()) #current position of the file cursor
file.seek(0) #move the file cursor to the beginning of the file
```

# Exception Handling

```python
try:
    file = open("data.txt", "r")
    content = file.read()
    print(content)
except FileNotFoundError:
    print("File not found.")
```
