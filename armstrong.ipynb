```python
num = int(input("Enter the number"))
original = num
digits = len(str(num))

sum_of_power = 0

while num>0:
    digit = num%10
    sum_of_power += digit ** digits
    num = num//10

if original == sum_of_power:
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
```
