```python
import seaborn as sns
sns.get_dataset_names() #gives the list of avaible datasets in seabord
```

```python
titanic = sns.load_dataset('titanic') #load the titanic dataset frome sebaborn
titanic.isnull().sum() #gives the count of null values in each column of the titanic dataset
```

- When a column has <40% missing values of the data, then we simply remove that column
- When a column has <40% missing values of data, then we handle those missing values means we replace those missing values with any number
    - here are two type of colums in any dataset - CATAGORICAL & NUMARICAL
    - When the column is numarical, we always replace the null values with the column mean(average)
    - When the column is categorical, we always replace the null values with mostfrequent(mode) values


```python
titanic.isnull().sum()
```

```python
titanic = titanic.dropna(subset=['embarked']) #drop the rows where the embarked column has nulll values
titanic.isnull().sum() #gives the count of null values in each column
```

```python
avg_age = round(titanic['age'].mean()) # calculate the average age of passengers in the titanic dataset
print(avg_age)
```

```python
#replace null values with avg_age
titanic['age'] = titanic['age'].fillna(avg_age) #replace null values in the 'age column with avg_age
titanic.isnull().sum()
```

```python
# checking duplicate values
titanic.duplicated()
```

### DATA MANIPULATION

```python
titanic
```

```python
#adding a new column
titanic = titanic.assign(Discount=titanic["fare"]*0.10)
titanic
```

```python
titanic['fare'] = titanic['fare'].apply(lambda x:x*1.05)  # hm kahi bhi new column add kr skte hai
titanic
```

```python
titanic["embarked"] = titanic["embarked"].map({
    "S" : "Southampton",
    "C" : "Cherbourg",
    "Q" : "Queentown",
})
titanic
```

```python
titanic.groupby("embarked")["fare"].mean()
```

```python
titanic.groupby("embarked")["fare"].max()
```

```python
titanic.groupby("sex")["fare"].min()
```

```python
titanic.groupby("sex")["fare"].max()
```
