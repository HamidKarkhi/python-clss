# 🔢 NumPy Data Science Cheat Sheet

---

## 1. Array Create Karna (Creation)
* **Code:** `np.array([1, 2, 3, 4, 5])`
  * **Use:** List ko 1D NumPy array mein badalne ke liye.
* **Code:** `np.zeros(5)`
  * **Use:** 5 elements ka khaali array (sab jagah 0) banane ke liye.
* **Code:** `np.ones(5)`
  * **Use:** 5 elements ka array (sab jagah 1) banane ke liye.
* **Code:** `np.full((2, 3), 7)`
  * **Use:** 2 rows aur 3 columns ka table jisme har jagah 7 likha ho.
* **Code:** `np.arange(1, 11, 2)`
  * **Use:** 1 se 10 tak ke numbers ka array ek-ek number chhod kar (1, 3, 5, 7, 9).
* **Code:** `np.eye(3)`
  * **Use:** 3x3 ki Identity Matrix banane ke liye (Machine Learning algorithms mein use hota hai).
* **Code:** `np.random.randint(10, 50, size=(3, 3))`
  * **Use:** Data science mein dummy data banane ke liye 10 se 50 ke beech ke random numbers ka 3x3 table.

---

## 2. Array Ki Details Check Karna (Inspection)
* **Code:** `arr.ndim`
  * **Use:** Array ki dimension check karne ke liye (1D hai, 2D hai ya 3D).
* **Code:** `arr.shape`
  * **Use:** Table mein kitne rows aur columns hain exact size janne ke liye.
* **Code:** `arr.size`
  * **Use:** Array ke andar total kitne elements (numbers) hain ginne ke liye.
* **Code:** `arr.dtype`
  * **Use:** Numbers ka data type check karne ke liye (int64 hai ya float64).

---

## 3. Shape Badatna (Reshaping)
* **Code:** `arr.reshape(1, 9)`
  * **Use:** Pura data ek single row mein convert karne ke liye.
* **Code:** `arr.flatten()`
  * **Use:** Kisi bhi 2D ya 3D matrix ko seedha karke 1D array banane ke liye (Deep Learning mein bahut use hota hai).

---

## 4. Specific Data Nikalna (Slicing & Indexing)
* **Code:** `a[1:4]`
  * **Use:** 1st index se lekar 3rd index tak ka data nikalne ke liye.
* **Code:** `b[0, 1]`
  * **Use:** 2D table mein se Pehli Row ka Doosra Column nikalne ke liye.
* **Code:** `b[1:3, 0:2]`
  * **Use:** Matrix mein se ek chota hissa (2nd aur 3rd row ka 1st aur 2nd column) kaat kar nikalne ke liye.

---

## 5. Maths Aur Statistics (Sabse Zyada Used)
* **Code:** `np.sum(a)` / `np.mean(a)`
  * **Use:** Array ka Total sum aur Average (Mean) nikalne ke liye.
* **Code:** `np.max(a)` / `np.min(a)`
  * **Use:** Sabse bada (Maximum) aur sabse chota (Minimum) number dhoodhne ke liye.
* **Code:** `np.std(a)` / `np.var(a)`
  * **Use:** Data mein kitna badlav (Standard Deviation aur Variance) hai check karne ke liye.
* **Code:** `np.sum(b, axis=0)`
  * **Use:** **Vertical Sum:** Har ek Column ka alag-alag total total nikalne ke liye.
* **Code:** `np.sum(b, axis=1)`
  * **Use:** **Horizontal Sum:** Har ek Row ka alag-alag total nikalne ke liye.
