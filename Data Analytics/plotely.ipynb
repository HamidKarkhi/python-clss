```python
import plotly.express as px
import seaborn as sns
```

```python
df = sns.load_dataset('car_crashes')
```

```python
#total accident by state
fig = px.bar(df, x='abbrev',y='total',
             title ='Statewise Car Accidents',
             labels = {'abbrev' : 'State', 'total' : 'Total Accidents'},
             
             )
fig.show()
```

```python
#total accident by state
fig = px.bar(df, x='abbrev',y='total',
             title ='Statewise Car Accidents',
             labels = {'abbrev' : 'State', 'total' : 'Total Accidents'},
             color = 'abbrev'
             )
fig.show()
```

```python
#total accident by state
fig = px.bar(df, x='abbrev',y='total',
             title ='Statewise Car Accidents',
             labels = {'abbrev' : 'State', 'total' : 'Total Accidents'},
             color = 'abbrev', template='plotly_dark'
             )
fig.show()
```

```python
#top 10 states  by insurence premium
top10 = df.nlargest(10, 'ins_premium') 
top10
```

```python
# Dataframe se top 10 rows nikalna
top10 = df.sort_values('ins_premium', ascending=False).head(10)

# Sahi template 'plotly_dark' use karke chart banana
fig = px.bar(top10, x='abbrev', y='ins_premium', 
             title="Top 10 states by insurance premium",
             template='plotly_dark', 
             color='ins_premium')

fig.show()
```

```python
#percentage of speeding accident (pai chart)
fig = px.pie(df, values = 'speeding', names='abbrev',
             title='Percentage of speeding accidents',
             template='plotly_dark',
             color_discrete_sequence=px.colors.sequential.Blackbody_r)
fig.update_traces(textposition='inside')
fig.show()
```
