# Break & Continue

```python
fruits = ['apple','mango','banana','cherry','tomato','guava']
for x in fruits:
    if x == "tomato":
        break
    print(x)
```

```python
fruits = ['apple','mango','banana','cherry','tomato','guava']
for x in fruits:
    if x == "tomato":
        continue
    print(x)
```

```python
#WAP to check whether a number is prime or not

num = int(input("Enter a number:"))
is_prime = True

if num <= 2:
    is_prime = False
else:
    for i in range(2,num):
        if num % i ==0:
            is_prime = False
            break
if is_prime:
    print("Prime Number")
else:
    print("Not a prime number")
```

```python
#Reverse a number
num = int(input("Enter a Number"))
reverse = 0

while num>0:
    digit = num%10
    reverse = reverse*10+digit
    num = num//10
```

```python
#Reverse a number
num = int(input("Enter a number"))
orig = num
reverse = 0

while num>0:
    digit = num%10
    reverse = reverse*10+digit
    num = num//10

print("Original Number:", orig)
print("Reversed Number:", reverse)
```
